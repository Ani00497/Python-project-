```python
#12.	Write a program in Python to check a triangle is equilateral,scalene or isosclees..
def check_triangle_type(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return "Invalid side lengths"
    if a + b <= c or a + c <= b or b + c <= a:
        return "Not a triangle"
    if a == b == c:
        return "Equilateral"
    elif a == b or b == c or a == c:
        return "Isosceles"
    else:
        return "Scalene"
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))

result = check_triangle_type(a, b, c)
print(f"The triangle is: {result}")
```

```python
#13.	Write a program in Python to check a given year is leapyear or not.
# Input the year to check
year = int(input("Enter a year: "))
if (year % 4 == 0):
    if (year % 100 == 0):
        if (year % 400 == 0):
            print(f"{year} is a leap year.")
        else:
            print(f"{year} is not a leap year.")
    else:
        print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
#16.	Write a program in Python to find the roots of Quadratic equation.
import math
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))
discriminant = b**2 - 4*a*c


if discriminant > 0:
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"The equation has two distinct real roots: {root1:.2f} and {root2:.2f}")
elif discriminant == 0:
    root = -b / (2*a)
    print(f"The equation has two identical real roots: {root:.2f}")
else:
    real_part = -b / (2*a)
    imaginary_part = math.sqrt(-discriminant) / (2*a)
    print(f"The equation has two complex roots: {real_part:.2f} + {imaginary_part:.2f}i and {real_part:.2f} - {imaginary_part:.2f}i")
```

```python
# 21.	Write a program in Python to print Reverse of a Numbers

number = int(input("Enter a number: "))
reversed_number = 0
while number > 0:
    digit = number % 10
    reversed_number = reversed_number * 10 + digit
    number = number // 10
print("Reverse of the number is:", reversed_number)
```

```python
# 24.	Write a Python program to check a number is prime or not.

number = int(input("Enter a number: "))
is_prime = True

if number < 2:
    is_prime = False
else:
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
#  25.	Write a Python program to check a number is Armstrong or not.

number = int(input("Enter a number: "))
digits = str(number)
num_digits = len(digits)
sum_of_powers = 0
for digit in digits:
    sum_of_powers += int(digit) ** num_digits
if sum_of_powers == number:
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")

```

```python
# 31.	Write a Python program to perform Linear search.

numbers = [3, 5, 7, 2, 8, 1]
target = 7

found = False
for i in range(len(numbers)):
    if numbers[i] == target:
        found = True
        print(f"Element {target} found at index {i}.")
        break

if not found:
    print(f"Element {target} not found in the list.")
```

```python
# 34.	Write a program that computes the net amount of a bank account based a transaction log from console input. The transaction log format is shown as following:
#D 100
#W 200
#¡
#D means deposit while W means withdrawal.
#Suppose the following input is supplied to the program:
#D 300
#D 300
#W 200
#D 100
#Then, the output should be:
#500

net_amount = 0
print("Enter transactions (D for Deposit, W for Withdrawal). Type 'done' to finish:")

while True:
    transaction = input()
    if transaction.lower() == 'done':
        break


    action, amount = transaction.split()
    amount = int(amount)

    if action == 'D':
        net_amount += amount
    elif action == 'W':
        net_amount -= amount
    else:
        print("Invalid transaction!")
print("Net amount:", net_amount)
```
